#### Data Dictionary - House Data

**id**:
> Description:10 character number that is unique to each house that allows houses to be tracked and identified specifically.
>
> Data type:Integer

**bedrooms**:
> Description:Number that shows how many bedrooms are in each house(usually single digits).
>
> Data type:Integer

**bathrooms**:
> Description:Decimal that shows how many bathrooms are in each house. This number is sometimes not an integer, and may have a decimal attached.
>
> Data type:Decimal

**sqft_living**:
> Description:Integer that shows the square footage of space that is used for the living space (not the entire amount of land, just the house space). 
>
> Data type:Integer

**sqft_lot**:
> Description:This is a number that shows the total square footage of the lot of the house. This number will have a decimal attached.
>
> Data type:Float

**floors**:
> Description:This is a decimal data type that shows how many floors the house includes. This number will be in single digits. It is possible that the numbers may not be integers.
>
> Data type:Float

**waterfront**:
> Description:This is a boolean value that shows if this house has a waterfront or not. Yes if true, No if not.
>
> Data type:Boolean

**view**:
> Description:This is a string that explains what view the house has. Some values may include Very Poor, Neutral, or Great.
>
> Data type:String

**condition**:
> Description:This is a string that explains what condition the house is in. Some values may include OK Condition, Good Condition, or Great condition.
>
> Data type:String

**grade**:
> Description:Grade is an number that grades how good the house is. The grading is in a 1 - 10 scale, with 10 being a great grade and 1 being a very poor grade.
>
> Data type:Float

**sqft_above**:
> Description:This is a number that shows the square footage above ground level.
>
> Data type:Float

**sqft_basement**:
> Description:This is an integer that shows the square footage of the basement included with the house. It must be noted that some houses do not have basements, so 0 square feet would be shown if that is the case.
>
> Data type:Integer

**yr_built**:
> Description:This is a number that shows what year the house was built. 
>
> Data type:Float

**yr_renovated**:
> Description:This is a number that shows the year the house was renovated. It must be noted that house may not have had a renovation, and an NaN value may be in place.
>
> Data type:Object

**zipcode**:
> Description:This is a five digit number that shows the zipcode the house is located in.
>
> Data type:Float

**lat**:
> Description:This is a decimal number that shows the latitude of the location of the house. Numbers may be negative as well as non-negative.
>
> Data type:Decimal

**long**:
> Description:This is a decimal number that shows the longitude of the location of the house. Numbers may be negative as well as non-negative.
>
> Data type:Decimal

**sqft_living15**:
> Description:This is an integer that shows the square footage of interior living space for the nearest 15 neighbors relative to the house.
>
> Data type:Integer

**sqft_lot15**:
> Description:This is an integer that shows the square footage of lot space for the nearest 15 neighbors relative to the house.
>
> Data type:Integer

#### Data Dictionary - House Sales Data

**id**:
> Description:10 character number that is unique to each house that allows houses to be tracked and identified specifically.
>
> Data type:Integer

**date**:

**price**:

### Step 2: Treat Names / Casing / Spelling Issues

```python
######################### TO DO ###################################

####### FOLLOW PROJECT DIRECTIONS TO COMPLETE THIS STEP ###########
```

```python
### DEFINE ###
# Drop unnamed 0 column in house_df.

### CLEAN ### 
house_df.drop(columns = ['Unnamed: 0'], inplace = True)
data_df.drop(columns = ['Unnamed: 0'], inplace = True)

### TEST ###
house_df.head() # Successful output should not show unnamed 0 column.
data_df.head()  # Successful output should not show unnamed 0 column.
```

```python
### DEFINE ###
# Turn columns in house_df into lowercase.

### CLEAN ### 
house_df = house_df.rename(columns = str.lower)

### TEST ###
house_df.head() # Columns should now be in lowercase
```

```python
### DEFINE ###
# Turn columns in data_df into lowercase.

### CLEAN ### 
data_df = data_df.rename(columns = str.lower)

### TEST ###
data_df.head() # Columns should now be in lowercase
```

```python
### DEFINE ###
# Remove punctuation and spaces from column names in data_df and convert to snake case.

### CLEAN ###
import regex as re
column_list = data_df.columns.to_list()
new_columns = []

for column in column_list:
    column = column.replace(' ', '_')
    column = column.replace('?', '')
        
    new_columns.append(column)

data_df.columns = new_columns

### TEST ###
data_df.head()
```

```python
### DEFINE ###
# Remove $, and commas from price in house_df

### CLEAN ### 
house_df.price = house_df.price.str.replace('$', '')
house_df.price = house_df.price.str.replace(',', '')

### TEST ###
house_df.head() # Dollar sign and comma should no longer be there
```

```python
### DEFINE ###
# Convert price from house_df into a str

### CLEAN ### 
house_df['price'] = house_df.price.astype('float')

### TEST ###
house_df.info() # price should read as float instead of object.
```

```python
### DEFINE ###
# Change ID datatype to a string

### CLEAN ### 
house_df['id'] = house_df.id.astype('str')

### TEST ###
house_df.info() # id should now read as an object instead of an integer.
```

```python
### DEFINE ###
# Change date data type from string to datetime as well as trim blank timestamps.

### CLEAN ### 
house_df['date'] = house_df['date'].str.replace('T000000', '')
house_df['date'] = pd.to_datetime(house_df['date'], format='%Y%m%d')

### TEST ###
house_df.info()
house_df.head() # date data type should read datetime and blank timestamps should be trimmed.
```

```python
### DEFINE ###
# Change ID data type from integer to string (object) in data_df.

### CLEAN ### 
data_df['id'] = data_df.id.astype('str')

### TEST ###
data_df.info() #Id data type should read object.
```

```python
### DEFINE ###
# Change zipcode data type from integer to string (object) in data_df.

### CLEAN ### 
data_df['zipcode'] = data_df.zipcode.astype('str')

### TEST ###
data_df.info() #zipcode data type should read object.
```
